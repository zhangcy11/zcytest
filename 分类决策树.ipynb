```python
#获取鸢尾花iris数据集
import pandas as pd
from sklearn import datasets
X,y=datasets.load_iris(return_X_y=True) #X与y
target_names=datasets.load_iris().target_names #y的值列表:0:setosa,1:versicolor,2:virginica
feature_names=datasets.load_iris().feature_names #特征X的名称列表
```

```python
print(target_names)
print(feature_names)
print(X[0:5])
print(y[0:5])
```

```python
#使用sklearn中的tree模块,设定max_depth=3(最大树深度为3)
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X, y)
```

```python
#预测sepal length=6,sepal width=1,petal length=3,petal width=1最可能是什么类型的花
clf.predict([[6.,1.,3.,1.]])
```

```python
#预测这个花属于各个类别的概率
clf.predict_proba([[6.,1.,3.,1.]])
```

```python
#使用treeplot画出决策树结构
from matplotlib import pyplot as plt
import sklearn
from sklearn.tree import plot_tree
plt.figure(figsize=(10,5))
tree.plot_tree(clf)
plt.show()
```

```python
#当X[2] <= 2.45时，预测结果为类别0，即setosa；
#当2.45 < X[2] <=4.95且X[3] <= 1.75时，预测结果为类别1，即versicolor；
#当4.85 < X[2]且1.75 < X[3]时，预测结果为类别2，即virginica;
```
